## Additional examples

### Two node network

```python
net_spec = """v1 : v1 + v2  : E
              v2 : ~v1 + v2 : E"""

network = DSGRN.Network(net_spec)

DSGRN.DrawGraph(network)
```

```python
parameter_graph = DSGRN.ParameterGraph(network)

print('Number of parameter nodes:', parameter_graph.size())
```

Compute isomorphism classes and make a dictionary to color the parameter graph nodes.

```python
%%time

# Isomorphism classes
iso_classes = DSGRN.Blowup.IsomorphismQuery(network)

# Create isomorphic node classes to color the nodes
node_classes = {k: iso_classes[k] for k in range(len(iso_classes))}
```

```python
bool_params = DSGRN.boolean_parameters(parameter_graph)

# Draw Boolean nodes as squares and the other nodes as circles
node_shapes = {v: 'square' if v in bool_params else 'circle' for v in range(parameter_graph.size())}
```

Draw the parameter graph. The nodes are color coded according to the Morse graph isomorphism class. Boolean nodes are shown as squares and the remaining nodes are shown as circles.

```python
DSGRN.draw_parameter_graph(parameter_graph, node_classes=node_classes, node_shapes=node_shapes, cmap=matplotlib.cm.tab20)
# DSGRN.draw_parameter_graph(parameter_graph, node_classes=node_classes, node_shapes=node_shapes)
```

Create a color coded legend for the isomorphim classes

```python
cmap=matplotlib.cm.tab20
max_clr_index = max(node_classes.keys())
cmap_norm = matplotlib.colors.Normalize(vmin=0, vmax=max_clr_index)

edges = [(0, 6), (1, 7), (2, 8), (3, 9), (4, 10), (5, 11)]

def class_label(k):
    # Add one extra spacing to separator for short labels
    sep = ':  ' if len(node_classes[k]) < 10 else ': '
    return 'Class ' + str(k) + sep + str(len(node_classes[k])) + ' nodes'
    # return 'Class ' + str(k) + ': ' + str(len(node_classes[k])) + ' nodes'

def class_color(k):
    return matplotlib.colors.to_hex(cmap(cmap_norm(k)), keep_alpha=True)

def class_shape(k):
    if node_classes[k].intersection(bool_params):
        return 'rectangle'
    return 'box'

def class_style(k):
    if node_classes[k].intersection(bool_params):
        return 'filled'
    return 'filled,rounded'

graphviz_str = 'graph {' + '\n' + \
               'graph [nodesep=0.1; ranksep=0.1];' + '\n' + \
               'edge [style=invis];' + '\n' + \
               'node [margin=0.05; height=0; width=0];' + '\n' + \
               '\n'.join(['"' + str(k) + '" [label="' + class_label(k) + '"; shape="' + class_shape(k) + '";' + \
               'style="' + class_style(k) + '";' + 'fontsize=12; fillcolor="' + class_color(k) + '"];' for k in node_classes]) + '\n' + \
               '\n'.join(['"' + str(k1) + '" -- "' + str(k2) + '";' for k1, k2 in edges ]) + '\n' + '}\n'

graphviz.Source(graphviz_str)
```

Show Morse graphs for each one of the classes

```python
for k, iso_class in enumerate(iso_classes):
    class_nodes = list(iso_class)
    print('Class', k, ':', len(class_nodes), 'parameter nodes')
    par_index = class_nodes[0]
    parameter = parameter_graph.parameter(par_index)
    morse_graph, stg, graded_complex = DSGRN.Blowup.ConleyMorseGraph(parameter)
    graph = DSGRN.Blowup.PlotMorseGraph(morse_graph)
    display(graph)
    print()
```

Define the essential Boolean parameter graph

```python
bool_net_spec = """x : (~y)(~z) : B E
                   y : (~x)(y) : B E
                   z : (~y)(z) : B E"""

bool_network = DSGRN.Network(bool_net_spec)

bool_parameter_graph = DSGRN.ParameterGraph(bool_network)

print(bool_parameter_graph.size())
```

Get list of indices of Boolean parameters embedded in the parameter graph of the original network

```python
# List of Boolean parameter indices
bool_parameters = []
for bool_pindex in range(bool_parameter_graph.size()):
    # Get the Boolean parameter
    bool_par = bool_parameter_graph.parameter(bool_pindex)
    # Get its index in the original parameter graph
    full_pindex = parameter_graph.index(bool_par)
    # Add the index to the list of Boolean parameters
    bool_parameters.append(full_pindex)

bool_parameters = set(bool_parameters)

print('Number of Boolean parameter nodes:', len(bool_parameters))
```

Count number of parameter nodes according to stability and periodicity

```python
%%time

stability_counts = defaultdict(int)
stability_counts_bool = defaultdict(int)
num_periodic = 0
num_periodic_bool = 0

for par_index in range(parameter_graph.size()):
    parameter = parameter_graph.parameter(par_index)
    morse_graph, stg, graded_complex = DSGRN.Blowup.ConleyMorseGraph(parameter)
    stable_nodes = [node for node in morse_graph.vertices() if len(morse_graph.adjacencies(node)) == 0]
    num_stable = len(stable_nodes)
    # Increment stability counter
    stability_counts[num_stable] += 1
    if par_index in bool_parameters:
        stability_counts_bool[num_stable] += 1
    # Search for stable periodic orbit
    for node in morse_graph.vertices():
        conley_index = morse_graph.vertex_label(node)[2]
        # Check if index of stable periodic orbit
        if conley_index == (1, 1, 0, 0):
            num_periodic += 1
            if par_index in bool_parameters:
                num_periodic_bool += 1
            # print(par_index, conley_index, num_stable)
```

```python
parameter_graph = DSGRN.ParameterGraph(network)

print('Number of parameter nodes:', parameter_graph.size())
```

```python
net_spec = """TGFb   : (~miR200)(~Ovol2) : M E
              miR200 : (~Snail1)(~Zeb1) : M E
              Snail1 : (~Snail1)(~miR34a)(TGFb) : M E
              Ovol2  : (~Zeb1) : M E
              Zeb1   : (~miR200)(~Ovol2)(Snail1) : M E
              miR34a : (~Snail1)(~Zeb1) : M E"""

network = DSGRN.Network(net_spec)

ess_parameter_graph = DSGRN.ParameterGraph(network)

print('Number of essential parameter nodes:', ess_parameter_graph.size())
```

```python
par_index = 7314075446

parameter = ess_parameter_graph.parameter(par_index)
```

```python
%%time
morse_graph, stg, graded_complex = DSGRN.Blowup.ConleyMorseGraph(parameter)
```

```python
DSGRN.Blowup.PlotMorseGraph(morse_graph)
```

Print cells for each stable Morse graph node

```python
# Get list of stable Morse graph vertices
stable_nodes = [node for node in morse_graph.vertices() if len(morse_graph.adjacencies(node)) == 0]

for v in stable_nodes:
    # Get Morse node coresponding to vertex
    morse_node = morse_graph.vertex_label(v)[0]
    # Get recurrent component corresponding to Morse node
    verts, edges = DSGRN.Blowup.MorseComponent(morse_graph, stg, graded_complex, morse_node, blowup=False)
    if len(verts) > 1:
        print('Morse node', morse_node, ': Multiple cells')
        continue
    cell = list(verts.values())[0]
    print('Morse node', morse_node, ':', cell)
```
